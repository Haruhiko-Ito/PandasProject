# Data Cleaning

### Cleaning Missing Values

```python
# For stumps & dead trees, going to change 'health' with Not Applicable, rather than NaN
mask = ((tree_census_subset['status'] == 'Stump') | (tree_census_subset['status'] == 'Dead'))
```

```python
# Now that 'mask' specifies all the rows that we want to change, let's specify the rows
# In the 'loc', the first element specifies the rows (mask) and the second element specifies the column

## tree_census_subset.loc[mask,'health'] = tree_census_subset.loc[mask,'health'].fillna('Not Applicable')
# This would fill all the 'health' that has NaN to 'Not Applicable'


# This will fill all columns with NaN to 'Not Applicable'
tree_census_subset.loc[mask] = tree_census_subset.loc[mask].fillna('Not Applicable')
```

```python
tree_census_subset[tree_census_subset['status'] == 'Stump']
```

```python
# Let's check if there are any other values that are missing
tree_census_subset.isna().sum()
```

```python
tree_census_subset[tree_census_subset['health'].isna()]

# Use this code and replace 'health' with all the other columns that still have NaN
```

```python
# Based on our observations of the NaN values that are left, here are our conclusion regarding how to deal with them

tree_census_subset['health'].fillna('Good', inplace=True)
tree_census_subset['problems'].fillna('None', inplace=True)
tree_census_subset['spc_latin'].fillna('No Observation', inplace=True)
```

```python
# Now there are no NaN values
tree_census_subset.isna().sum()
```

### Dealing with Outliers

```python
big_tree = tree_census_subset[ (tree_census_subset['tree_dbh']>60) | (tree_census_subset['stump_diam']>60) ]
big_tree
```

```python
# We can either update our exisintg tree_census_subset or create another subset (here we're going to create another subset)
# We are creating a new subset w/out the outliers so that we can calculate sensible statistical values
# Note that we are now using & instead of |
# This is because we need both columns to not be outliers

tree_census_no_outlier = tree_census_subset[ (tree_census_subset['tree_dbh']<=60) & (tree_census_subset['stump_diam']<=60) ]
tree_census_no_outlier
```

```python
# Now we are going to separate the 'Alive' trees vs. 'Stump & Dead' trees
tree_census_subset_alive = tree_census_subset[tree_census_subset['status'] == 'Alive']
tree_census_subset_dead_or_stump = tree_census_subset[ (tree_census_subset['status'] == 'Stump') |
                                                        (tree_census_subset['status'] == 'Dead') ]
```

##### For Alive Trees

```python
# Look at satistical values based on tree type
tree_census_subset_alive.groupby('spc_latin')['tree_dbh'].describe()
```

```python
# Here we can see that there are values that are simply too small & too large
# We are going to assume that any tree below 25th percentile is too small --> replace all those values to the 25th percentile
# 75th percentile and above are too large --> replace w/ 75th percentile

# We are going to take 'stats' as a reference
# List only the 25th & 75th percentile

stats_alive = tree_census_subset_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin','25%','75%']]
stats_alive
```

```python
# Now merge stats_alive with the entire alive table
tree_census_subset_alive_merged = tree_census_subset_alive.merge(stats_alive, on='spc_latin', how='left')
tree_census_subset_alive_merged
```

```python
# Now apply our assumption
mask2 = tree_census_subset_alive_merged['tree_dbh'] < tree_census_subset_alive_merged['25%']
tree_census_subset_alive_merged.loc[mask2, 'tree_dbh'] = tree_census_subset_alive_merged['25%']

mask3 = tree_census_subset_alive_merged['tree_dbh'] > tree_census_subset_alive_merged['75%']
tree_census_subset_alive_merged.loc[mask3, 'tree_dbh'] = tree_census_subset_alive_merged['75%']
```
