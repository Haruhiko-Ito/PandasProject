# 1. Data Cleaning

### 1.1 Data Types
Check dtypes and see if they are suitable for corresponding columns

```python
df.dtypes
```

Most dtypes are suitable. Although 'votes', 'budget', 'gross', and 'runtime' do not require decimal precision, we cannot change dtype from 'float' to 'int' due to presence of null values. Therefore, we will keep the current dtype as this will not affect data.

We can see that the 'released' column is currently holding 2 different information: 1) the released date and 2) the released location. Therefore, we are going to split this column into two: 1) released_date and 2) released_location

```python
y = df2['released'].str.replace(")", "").str.split("(", expand=True).rename(columns = {0:'released_date', 1:'released_location'})
y
```

```python
# Since 'released' column was the 5th column, we are going to insert 'released_date' & 'released_location' correspondingly
df2.insert(4, 'released_date', y['released_date'])

df2.insert(5, 'released_location', y['released_location'])
```

```python
df2.drop(['released'], axis=1, inplace=True)
df2
```

```python
# Convert 'released_date' into dtype 'datetime'

df2['released_date'] = df2['released_date'].astype(object).astype('datetime64')
df2
```

### 1.2 Dealing with Missing Values

```python
# Checking which columns have null values
df2.isna().sum()
```

We can see that there are 2 rows with null values for 'released_location' & 'released_date'.

Check to see that these 2 rows both represent the null values for the columns

```python
df2[df2['released_date'].isna()]
```

Since there are only 2 rows with null values, let's search the information and add them

```python
# Replacing 'released_location' values

mask1 = df2['name'] == 'Saw: The Final Chapter'
df2.loc[mask1, 'released_location'] = 'United States'

mask2 = df2['name'] == 'The Wolfman'
df2.loc[mask2, 'released_location'] = 'United States'
```

```python
# Replacing 'released_date' values

df2.loc[mask1, 'released_date'] = '2010-10-29'
df2.loc[mask2, 'released_date'] = '2010-02-12'
```

```python
# Check to see that there are no null values for 'released_location' & 'released_date'
df2[['released_date','released_location']].isna().sum()
```

For the remaining columns with null values, the following are our decision protocol for null values.
1. dtype(float64): 'score' 'votes' 'budget' 'gross' 'runtime'

'runtime', 'score', 'votes': Fill these columns with 0

'gross', 'budget': Fill these columns np.mean

2. dtype(object): 'rating', 'writer', 'star', 'country', 'company'

```python
columns_1 = ['rating','company','writer','country','star']
for i in columns_1:
    df2[i].fillna("Others",inplace=True)

columns_2 = ['runtime','score','votes']
for j in columns_2:
    df2[j].fillna(0,inplace=True) 

    columns_3 = ['budget','gross']
for k in columns_3:
    df2[k].fillna(round(np.mean(df2[k])),inplace=True)
```

```python
# Check to make sure null values are not present
df2.isna().sum().sort_values(ascending=False)[:-4].reset_index().rename(columns={'index':'column_name', 0:'Null Values'})
```

### 1.4 Data Values

In the table, it seems that the 'year' column is not correspondent to the year of the 'released_date' column. Therefore, we are going to create a new year column that has the correct years

```python
# Check that 'year' column doesn't correspond to the years of 'released_date'

df_temp = df2.assign(year_check = 
                    df2['released_date'].dt.year == df2['year'])
df_temp['year_check'].value_counts()
```

```python
# Since there is 1324 cases when 'year' column is different from year of 'released_date', create new column 'correctyear'
df2['yearcorrect'] = df2['released_date'].dt.year
df2
```

Note, if we simply wanted to replace the 'year' column with the years of the 'released_date' column, we can use the following code:

df2['year'] = df2['released_date'].dt.year
